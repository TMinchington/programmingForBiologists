# For loops

- For loops let us loop over iterable objects.
- They can be used to repeat the same action over and over again.


### Basic stucture:

for x in iterable:  <- This strucutre basically says generate variable x and each loop make it equal to something in an interable. The colon ':' needs to be after this or the for loop wont work. 
    print(x)        <- note that this line is indented. This is important as it tells python that this is inside the loop

```python
list1 = ["Like", "and", "subscribe"]
```

Remember that we can access parts of a list by their index

```python
print(list1[0]) # first element of the list
print(list1[1]) # second element of the list
print(list1[2]) # third element of the list
```

If you wanted to print every element in the list you could do this manually but this takes a long time.

### Option 1: Use a for loop to cycle through the index of a list
(not the best option in python, but might be your only option in other programming languages)

```python
for x in range(0, len(list1)):

    print(x) # this would print your index
```

So what is actually happening here?

range() is a function that basically generates all the values better a start number and an end number. So above the start is '0' and the end in len(list). 

```python
print(len(list1)) # len() is a function that returns the length of certain objects. Here is will give us the length of a list.
```

We can see what range is doing by converting it to a list. Which is kind of what happens inside the for loop.

```python
list(range(0, len(list1)))
```

Ok so lets use this to loop through our list 1

```python
for x in range(0, len(list1)):

    print(x) # this would print your index
    print(list1[x]) # here we use the index generated by our loop to print each item in the list
```

### Option 2: Loop through the list directly

In python you can loop over iterables directly, this is easier to read and less code. 

```python
for x in list1:
    print(x)
```

- You can loop over and iterable object. You can also nest loops (put on inside another).
- Therefore we can also loop the text in each item of list1



```python
for x in list1:
    for y in x:
        print(y) # here we print every letter in each word. 
```

```python
for x in list1:
    print(x) # you can also print the word in each of the outer loop
    for y in x:
        print(y) # you can see that this loop happens once for each word in the list
```

You don't need to use x and y these are basically temporary variable that you make in side the loop

```python
for x in list1:
    print(x) # you can also print the word in each of the outer loop
    for y in x:
        print(y) # here we print every letter in each word. 
```

Is the same as:

```python
for word in list1:
    print(word) # you can also print the word in each of the outer loop
    for letter in word:
        print(letter) # here we print every letter in each word. 
```
